```python
import pandas as pd
import numpy as np
import random

# 1. Base Data Lists
event_types = ['Naissance', 'Geburt', 'Mariage', 'Heirat', 'Décès', 'Tod', 'Birth', 'Death']
cities = ['Paris', 'Berlin', 'Lyon', 'Munich', 'Marseille', 'Hamburg', 'Strasbourg']
first_names = ['Jean', 'Hans', 'Marie', 'Anna', 'Pierre', 'Klaus', 'Sophie', 'Lukas', 'Claire', 'Julia']
last_names = ['Martin', 'Müller', 'Bernard', 'Schmidt', 'Dubois', 'Schneider', 'Roux', 'Fischer']

# Messy Date Formats to randomly choose from
date_formats = ["%d-%m-%Y", "%Y/%m/%d", "%d.%m.%y", "%d-%b-%Y"]

data = []

# 2. Generate 500 records
for i in range(500):
    # Pick random details
    e_type = random.choice(event_types)
    city = random.choice(cities)
    
    # Mess up the names intentionally (mixed casing, extra spaces)
    f_name = random.choice(first_names)
    l_name = random.choice(last_names)
    if random.random() > 0.5:
        name = f"{f_name.lower()} {l_name.upper()}"
    elif random.random() > 0.8:
        name = f"  {f_name}   {l_name}  " # Extra spaces
    else:
        name = f"{f_name} {l_name}"

    # Generate a random date in 1990s
    year = random.randint(1990, 1999)
    month = random.randint(1, 12)
    day = random.randint(1, 28)
    clean_date = pd.Timestamp(year, month, day)
    
    # Format the date messily
    messy_date = clean_date.strftime(random.choice(date_formats))
    
    # Introduce French/German specific month abbreviations randomly
    if "May" in messy_date: messy_date = messy_date.replace("May", "Mai")
    if "Dec" in messy_date: messy_date = messy_date.replace("Dec", "Déc")

    data.append([f"CERT-{1000+i}", e_type, messy_date, name, city])

# 3. Create DataFrame
df = pd.DataFrame(data, columns=['Record_ID', 'Event_Type', 'Event_Date', 'Primary_Name', 'City'])

# 4. Inject Missing Values (Nulls)
# 5% of names missing, 10% of cities missing
df.loc[df.sample(frac=0.05).index, 'Primary_Name'] = np.nan
df.loc[df.sample(frac=0.10).index, 'City'] = np.nan

# 5. Inject Duplicates (Copy 20 random rows and append them)
duplicates = df.sample(n=20)
df = pd.concat([df, duplicates], ignore_index=True)

# 6. Shuffle the dataset to mix the duplicates in
df = df.sample(frac=1).reset_index(drop=True)

# 7. Export to CSV
df.to_csv('raw_messy_certificates.csv', index=False)

print("✅ Success! 'raw_messy_certificates.csv' has been generated with", len(df), "rows.")
```
